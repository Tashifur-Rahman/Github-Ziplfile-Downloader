```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
df=pd.read_csv('/content/california_housing.csv')
df.head(5)
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.columns
```

```python
features=df[['longitude','latitude','housing_median_age','total_rooms','total_bedrooms','population','households','median_income','ocean_proximity']]
x=pd.get_dummies(features,drop_first=True)
y=df[['median_house_value']]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
y.head(5)
```

```python
x.head(5)
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
```

```python
print(prediction[:5])
```

```python
print(y_test.head(5))
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print("R2 score=",r2_score(y_test,prediction))
print("Average_error=",mean_squared_error(y_test,prediction))
```
